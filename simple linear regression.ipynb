```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('C:/Users/DEEPA/Desktop/regression exam/Salary_Data.csv')
```

```python
data.head()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

```python
y
```

## dividing data to train and test

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

## fitting simple linear regression to the training set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

#the amount of change in Ytrain(dependent) due to  1 unit change in X-train (independent )
##here we created the machine called regressor which is simple linear regression  and it learned on the training set here so that
the machine here understand the correlations between salary and years of experience and use this learning to predict the salry
based on its experience
###this is the most simple machine learning model

## predicting the test set results

```python
y_pred=regressor.predict(X_test)
y_pred
```

##the salaries in y_test are the actual salaries of 10 employess and the salaries in the y_pred are the predicted salaries of 10 employees

```python
y_test
```

## visualizing the training set results

```python
plt.scatter(X_train,y_train,color = 'red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary VS Expereince(Training Set)')
plt.xlabel('Years OF Expereicne')
plt.ylabel('Salary')
plt.show()
```

##we can see clearly a linear dependecy
